```python
import numpy as np
from matplotlib import pyplot as plt

#
np.random.seed(114)

#data
data1 = np.random.normal(loc=0,   scale=1, size=10000)
data2 = np.random.normal(loc=0.5, scale=2, size=10000)

#data1
plt.hist(data1, bins=50, density=True, histtype='step', color='blue')
plt.title("Outline Histogram of N(0,1)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

#data2
plt.hist(data2, bins=50, density=True, histtype='step', color='green')
plt.title("Outline Histogram of N(0.5,2)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

#2D Histogram
plt.hist2d(data1, data2, bins=50,cmap='Blues')
plt.colorbar(label="Counts")
plt.title("2D Histogram of data1 vs data2")
plt.xlabel("data1 values")
plt.ylabel("data2 values")
plt.show()
```
